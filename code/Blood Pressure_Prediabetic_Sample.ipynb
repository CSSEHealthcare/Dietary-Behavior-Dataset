```python
def Sample_Generator(DemoAge):
    zeros = [0.0] * 44
    temp1 = ind_food_1_df_mod[ind_food_1_df_mod.SEQN.isin(DemoAge.SEQN)] 
    temp2 = ind_food_2_df_mod[ind_food_2_df_mod.SEQN.isin(DemoAge.SEQN)] 
    SampleAge = pd.read_csv ("X.csv")
    count = 0
    for ind in temp1.index: 
        if (str (temp1['SEQN'][ind]) + 'Day1') not in SampleAge.columns: SampleAge[str (temp1['SEQN'][ind]) + 'Day1'] = zeros
    for ind in temp1.index: 
        for ind2 in SampleAge.index:
            if temp1['DR1IFDCD'][ind] ==  SampleAge['Food Code'][ind2]:
                SampleAge[str (temp1['SEQN'][ind])+ 'Day1'][ind2] += float (temp1['DR1IGRMS'][ind]) / SampleAge ['Serving (g)'][ind2]
    for ind in temp2.index: 
        if (str (temp2['SEQN'][ind]) + 'Day2') not in SampleAge.columns: SampleAge[str (temp2['SEQN'][ind]) + 'Day2'] = zeros
    for ind in temp2.index: 
        for ind2 in SampleAge.index:
            if temp2['DR2IFDCD'][ind] ==  SampleAge['Food Code'][ind2]:
                SampleAge[str (temp2['SEQN'][ind])+ 'Day2'][ind2] += float (temp2['DR2IGRMS'][ind]) / SampleAge ['Serving (g)'][ind2]
    return SampleAge.round(1)  
```
