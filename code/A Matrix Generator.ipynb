# Method 2: Grouping all Egg foods and Grouping Cakes and Breakfast meals

```python
A_based_on_data = pd.read_csv ("A-matrix-based-on-data.csv")
for (columnName, columnData) in A_based_on_data.iteritems():
    if columnName !='Food Code' and columnName !='Name' and columnName !='Serving (g)':
        for ind in A_based_on_data.index:
            temp1 = ind_food_1_df_mod[(ind_food_1_df_mod.DR1IFDCD == A_based_on_data['Food Code'][ind])] 
            temp1[columnName] = temp1[columnName]/temp1['DR1IGRMS']
            temp1[columnName] = temp1[columnName] * A_based_on_data['Serving (g)'][ind]
            A_based_on_data[columnName][ind] = temp1[columnName].mean()
            if ind == 12:
                temp1 = ind_food_1_df_mod[(ind_food_1_df_mod.DR1IFDCD == A_based_on_data['Food Code'][12]) 
                                         |(ind_food_1_df_mod.DR1IFDCD == A_based_on_data['Food Code'][13])
                                         |(ind_food_1_df_mod.DR1IFDCD == A_based_on_data['Food Code'][14])
                                         ] 
                temp1[columnName] = temp1[columnName]/temp1['DR1IGRMS']
                temp1[columnName] = temp1[columnName] * A_based_on_data['Serving (g)'][ind]
                A_based_on_data[columnName][ind] = temp1[columnName].mean()
                A_based_on_data['Name'][ind] = 'Egg Meals'
            if ind == 19:
                temp1 = ind_food_1_df_mod[(ind_food_1_df_mod.DR1IFDCD == A_based_on_data['Food Code'][19]) 
                                         |(ind_food_1_df_mod.DR1IFDCD == A_based_on_data['Food Code'][20])
                                         |(ind_food_1_df_mod.DR1IFDCD == A_based_on_data['Food Code'][21])
                                         |(ind_food_1_df_mod.DR1IFDCD == A_based_on_data['Food Code'][22])
                                         ] 
                temp1[columnName] = temp1[columnName]/temp1['DR1IGRMS']
                temp1[columnName] = temp1[columnName] * A_based_on_data['Serving (g)'][ind]
                A_based_on_data[columnName][ind] = temp1[columnName].mean()
                A_based_on_data['Name'][ind] = 'Cakes/Biscuits/Pancakes'
            if ind == 32:
                temp1 = ind_food_1_df_mod[(ind_food_1_df_mod.DR1IFDCD == A_based_on_data['Food Code'][32]) 
                                         |(ind_food_1_df_mod.DR1IFDCD == A_based_on_data['Food Code'][35])
                                         ] 
                temp1[columnName] = temp1[columnName]/temp1['DR1IGRMS']
                temp1[columnName] = temp1[columnName] * A_based_on_data['Serving (g)'][ind]
                A_based_on_data[columnName][ind] = temp1[columnName].mean()
                A_based_on_data['Name'][ind] = 'Greens/Vegetables'
            if ind == 38:
                temp1 = ind_food_1_df_mod[(ind_food_1_df_mod.DR1IFDCD == A_based_on_data['Food Code'][38]) 
                                         |(ind_food_1_df_mod.DR1IFDCD == A_based_on_data['Food Code'][39])
                                         ] 
                temp1[columnName] = temp1[columnName]/temp1['DR1IGRMS']
                temp1[columnName] = temp1[columnName] * A_based_on_data['Serving (g)'][ind]
                A_based_on_data[columnName][ind] = temp1[columnName].mean()
                A_based_on_data['Name'][ind] = 'Butters/Oils'
            
A_based_on_data_reduced = A_based_on_data.drop ([13,14,20,21,22,35,39])
A_based_on_data_reduced.to_csv((r'C:\Users\farzi\OneDrive - Johns Hopkins\Research\Inverse Optimization\Data\Python-codes\A_based_on_data_groups.csv'))
A_based_on_data_reduced
```
